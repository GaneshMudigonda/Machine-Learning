```python
df=df.drop('Alley',axis=1)
df=df.drop('Id',axis=1)
```

```python
df['LotFrontage'].fillna((df['LotFrontage'].median()), inplace=True)
```

```python
df['MasVnrType'].fillna(df['MasVnrType'].mode()[0], inplace=True)
df['MasVnrArea'].fillna(df['MasVnrArea'].mode()[0], inplace=True)
df['BsmtQual'].fillna(df['BsmtQual'].mode()[0], inplace=True)
df['BsmtCond'].fillna(df['BsmtCond'].mode()[0], inplace=True)
df['BsmtExposure'].fillna(df['BsmtExposure'].mode()[0], inplace=True)
df['BsmtFinType1'].fillna(df['BsmtFinType1'].mode()[0], inplace=True)
df['BsmtFinType2'].fillna(df['BsmtFinType2'].mode()[0], inplace=True)
df['Electrical'].fillna(df['Electrical'].mode()[0], inplace=True)
df['GarageType'].fillna(df['GarageType'].mode()[0], inplace=True)
df['GarageFinish'].fillna(df['GarageFinish'].mode()[0], inplace=True)
df['GarageQual'].fillna(df['GarageQual'].mode()[0], inplace=True)
df['GarageCond'].fillna(df['GarageCond'].mode()[0], inplace=True)
```

```python
df['GarageYrBlt'].fillna(df['GarageYrBlt'].mode()[0], inplace=True)
```

```python
df=df.drop(['FireplaceQu','PoolQC','Fence','MiscFeature'],axis=1)
```

```python
df.info()
```

```python
df.describe()
```

```python
y=df['SalePrice']
X=df.drop('SalePrice',axis=1)
```

```python
nonnumeric=X.columns[X.dtypes=='object']
```

```python
nonnumeric
```

```python
X=pd.get_dummies(X,columns=nonnumeric,drop_first=True)
```

```python
type(X)
```

```python
X.shape
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.4,random_state=4)
```

```python
X.describe()
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
```

```python
X_train=sc.fit_transform(X_train)
```

```python
X_test=sc.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
model.score(X_train,y_train)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
from sklearn.linear_model import Ridge
```

```python
ridge=Ridge()
```

```python
model=GridSearchCV(ridge,param_grid={'alpha':[0.1,0.2,0.3,0.5,1,5,10,20,50,100,500,1000]},cv=3,return_train_score=True)
```

```python
model.fit(X_train,y_train)
```

```python
pd.DataFrame(model.cv_results_)[['param_alpha','mean_train_score','mean_test_score']]
```

```python
ridge=Ridge(alpha=500)
```

```python
ridge.fit(X_train,y_train)
```

```python
ridge.score(X_train,y_train)
```

```python
ridge.score(X_test,y_test)
```
